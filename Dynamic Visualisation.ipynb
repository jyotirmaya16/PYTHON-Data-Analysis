```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/singhrau0/Big-Data-Preprocessing/main/youtubesample.csv')
```

```python
#piechart
def piechart(a,b,c):
  p=df[df['Genre']==a][b].sum()
  q=df[df['Genre']==a][c].sum()
  r=[p,q]
  s=[b,c]
  plt.pie(x=r,labels=s,autopct='%1.1f%%')
  plt.show()
```

```python
#line chart
def linechart(a,b,c):
  sns.lineplot(x=b,y=c,data=df[df['Genre']==a])
  plt.show()
```

```python
#scatter plot
def scatterplot(a,b,c):
  sns.scatterplot(x=b,y=c,data=df[df['Genre']==a])
  plt.show()
```

```python
good=widgets.Dropdown(
      options=[piechart,scatterplot,linechart],
      value=linechart, #difult value
      description='Choose Cahrt:', #agg function
      disabled=False
  )
def call(k):
  l=widgets.Dropdown(
      options=df['Genre'].unique(),
      value='Entertainment', #difult value
      description='Genre:', #agg function
      disabled=False
  )
  j = widgets.Dropdown(
      options=['likes', 'comment_count','dislikes'],
      value='likes', #difult value
      description='Criteria1:', #agg function
      disabled=False,
      )
  m = widgets.Dropdown(
      options=['likes', 'comment_count','dislikes'],
      value='dislikes', #difult value
      description='Criteria2:', #agg function
      disabled=False,
      )
  a=l
  b=j
  c=m
  widgets.interact(k,a=l,b=j,c=m)
widgets.interact(call,k=good)
plt.show()
```
